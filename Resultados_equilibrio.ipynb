```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline  # Interpolación spline

# Leer los datos ignorando la primera fila (ajusta encoding si es necesario)
file_path = r"C:/Users/german.perez/OneDrive - Destinus NL BV/Escritorio/Personal/TFG/Fotos/6 - Resultados/resultados_equilibrio.txt"
df = pd.read_csv(file_path, sep='\s+', decimal='.', header=None, skiprows=1, encoding='latin1')

# Definir los nombres de las columnas
df.columns = ['Tiempo', 'Presión', 'Temperatura', 'Cp', 'Mm', 'Rg', 'Gamma', 'Vsound', 
              'N_KOH', 'N_K2CO3', 'N_K2O', 'N_CO', 'N_CO2', 'N_H2O', 'N_H2', 'N_H', 
              'N_OH', 'N_O', 'N_O2', 'N_N2']

# Convertir columnas numéricas a flotante
df = df.astype(float)

# Interpolación spline para suavizar
x_smooth = np.linspace(df['Tiempo'].min(), df['Tiempo'].max(), 300)  # Más puntos
spl = make_interp_spline(df['Tiempo'], df['Temperatura'], k=3)  # k=3 para spline cúbico
y_smooth = spl(x_smooth)

# Graficar curva suavizada
plt.figure(figsize=(10, 4))
plt.plot(x_smooth, y_smooth, color='blue', label='Presión (suavizada)')
plt.xlabel("Tiempo (s)")
plt.ylabel("Presión (bar)")
plt.title("Presión durante el ensayo")
plt.legend()
plt.grid()
plt.show()
```
